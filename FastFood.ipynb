```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("FASTFOOD.csv")
```

```python
df
```

```python
df.columns
```

```python
plt.figure(figsize=(10,5))
plt.pie(pd.value_counts(df.item_type),labels=['Fastfood','Beverages'],autopct='%.1f%%')
plt.title('Item Type')
plt.legend()
plt.show()
```

```python
sns.set(rc={'figure.figsize':(10,5)})
ax = sns.countplot(x='item_name',data=df,color='red')
for x in ax.containers:
    ax.bar_label(x)
plt.title('Item Name against count',fontsize =15)
```

```python
sns.set(rc={'figure.figsize':(10,5)})
ax = sns.countplot(x='item_price',data = df,color='purple')
for x in ax.containers:
    ax.bar_label(x)
plt.title('Plot of Item price against the count')
```

```python
x1 = df['item_price'].tolist()
y1 = df['order_id'].tolist()
y2 = df['quantity'].tolist()
y3 = df['transaction_amount'].tolist()
```

```python
plt.figure(figsize=(10,5))
plt.pie(pd.value_counts(df.received_by),labels=['Male','Female'],autopct='%.1f%%')
plt.title('Received By')
plt.legend()
plt.show()
```

```python
width =0.5
plt.figure(figsize = (10,5))
plt.bar(x1,y1,width)
plt.xlabel('Price')
```

```python
sns.displot(x='quantity',data=df,color='green')
```

```python
plt.figure(figsize=(10,5))
plt.pie(pd.value_counts(df.time_of_sale),labels=['Morning','Afternoon','Evening','Night','Midnight'],autopct='%.2f%%')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.pie(pd.value_counts(df.transaction_type),labels=['Cash','Online'],autopct='%.2f%%')
plt.legend()
plt.show()
```
